```python
# import libraries
import pandas as pd
import numpy as np
import seaborn as sns
# Implementation of matplotlib function
import matplotlib.pyplot as plt
import matplotlib.lines as lines
# read csv file
df = pd.read_csv("Medical_Cost_Insurance_MCI.csv")
#print(df.shape)
df.info() #print dataframe columns, no nullvalues, count, datatype of dataset
```

```python
# creating data frame
df_bmi = df[["bmi", "age"]]
```

```python
df_1 = df_bmi.groupby("age")["bmi"].mean()
```

```python
#creatiing new csv file
df_1.to_csv("Manipulation_MCI.csv")
```

```python
# read  new csv file
df_new = pd.read_csv("Manipulation_MCI.csv")
df_new
```

```python
# creating data frame
df_region = df[["region", "charges"]]
```

```python
df2 = df_region.groupby("region")["charges"].sum()
```

```python
#creatiing new csv file
df2.to_csv("Manipulation_MCI_Region.csv")
```
